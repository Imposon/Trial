```python
from groq import Groq
import re
from google.colab import userdata

api_key = userdata.get('Huppa')  #os.getenv("GROQ_API_KEY")
if not api_key:
    raise EnvironmentError(
        "GROQ_API_KEY is not set. Set it in your shell before running this notebook."
    )

client = Groq(api_key=api_key)
print("Groq client initialized")

Model = "openai/gpt-oss-120b"

```

```python
def generate_response(prompt , max_tokens=500):
    response = client.chat.completions.create(
        model=Model,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content

def extract_answer(reponse):
    #using regex get everything between <answer> and </answer>
    answer = re.search(r"<answer>(.*?)</answer>", reponse, re.DOTALL)
    if answer:
        return answer.group(1).strip()
    else:
        return "No answer found in the response."

def match_answer(response, expected_answer):
    extracted_answer = extract_answer(response)
    return extracted_answer == expected_answer, extracted_answer
```

## Task 1

### Objective
You are auditing an employee reimbursement claim.

Claim lines:
- Taxi: 380
- Client meal: 620
- Personal dessert: 140
- Parking: 55

Policy rules:
- Reimburse taxi and parking fully.
- Reimburse client meal up to a maximum of 500.
- Do not reimburse personal items.

Compute the final reimbursable amount.

### Output Requirement
Return only the final number inside XML answer tags.

Example format: `<answer>935</answer>`

### Your task
Write a prompt in the next cell so the model applies all policy rules correctly and returns only the tagged answer.

```python
# Write your prompt here
prompt = "Think you are an auditing employee and you have to audit employees reimbursement claim the claim lines are taxi:380(full), client meal: 620(capped at 500), personal dessert: 140(exclude), parking: 55(full) and follow these policy rules strictly reimburse taxi and parking fully , reimburse client meal upto maximum of 500 and do not reimbure personal items and return the final number inside XML answer tags for example format <answer>935</answer>"

# Do not change below this line.
response = generate_response(prompt)
expected_answer = "935"
correct, extracted_answer = match_answer(response, expected_answer)

print("=" * 60)
print("MODEL RESPONSE")
print("=" * 60)
print(response)
print("=" * 60)

if correct:
    print(f"[PASS] Extracted answer: {extracted_answer}")
else:
    print(f"[FAIL] Extracted answer: {extracted_answer}")
    print(f"Expected answer: {expected_answer}")
    assert False, "Task failed. Update your prompt and try again."
```
